```python
from PIL import Image
import torchvision.transforms.functional as TF
from transforms import transpose
import matplotlib.pyplot as plt
```

```python
image = Image.open('./fig.png').resize((512, 512))

plt.figure(figsize=(40, 20))
plt.subplot(2, 8, 1)
plt.imshow(image)
plt.subplot(2, 8, 2)
plt.imshow(TF.vflip(image))
plt.subplot(2, 8, 3)
plt.imshow(TF.hflip(image))
plt.subplot(2, 8, 4)
plt.imshow(TF.hflip(TF.vflip(image)))

image = transpose(image)

plt.subplot(2, 8, 5)
plt.imshow(image)
plt.subplot(2, 8, 6)
plt.imshow(TF.vflip(image))
plt.subplot(2, 8, 7)
plt.imshow(TF.hflip(image))
plt.subplot(2, 8, 8)
plt.imshow(TF.hflip(TF.vflip(image)))

image = Image.open('./fig.png').resize((512, 512))

plt.figure(figsize=(40, 20))
plt.subplot(2, 8, 9)
plt.imshow(image)
plt.subplot(2, 8, 10)
plt.imshow(TF.rotate(image, 90))
plt.subplot(2, 8, 11)
plt.imshow(TF.rotate(image, 180))
plt.subplot(2, 8, 12)
plt.imshow(TF.rotate(image, 270))

image = TF.vflip(image)

plt.subplot(2, 8, 13)
plt.imshow(image)
plt.subplot(2, 8, 14)
plt.imshow(TF.rotate(image, 90))
plt.subplot(2, 8, 15)
plt.imshow(TF.rotate(image, 180))
plt.subplot(2, 8, 16)
plt.imshow(TF.rotate(image, 270))
```
