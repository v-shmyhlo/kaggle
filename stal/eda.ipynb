```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
areas = np.load('./stats.npy')
```

```python
weights = np.log(np.e + areas)
# weights = weights / weights.sum()
print(weights.min(), weights.max(), weights.sum())
plt.scatter(areas, weights, s=1)
```

```python
image = np.array(Image.open('../tmp.png').convert('L'))
plt.imshow(image)
plt.show()

x = image.sum(1) / image.sum(1).max()
plt.plot(x)
x = x > x.max() / 4
d = np.diff(x)
plt.plot(d)
plt.show()
l = np.where(d[:len(d)//2])[0][-1]
r = np.where(d[len(d)//2:])[0][0] + len(d) // 2

print(l, r)
image = image[l:r]
plt.imshow(image)
plt.show()

x = image.sum(0) / image.sum(0).max()
plt.plot(x)
x = x > x.max() / 4
d = np.diff(x)
plt.plot(d)
plt.show()
l = np.where(d[:len(d)//2])[0][-1]
r = np.where(d[len(d)//2:])[0][0] + len(d) // 2

print(l, r)
image = image[:, l:r]
plt.imshow(image)
plt.show()
```
